```python
from skimage import data, exposure, img_as_float
import numpy as np
import matplotlib.pyplot as plt

image = img_as_float(data.camera())
hist_img =exposure.histogram(image,10)

fig=plt.figure()
plt.subplot(111)
plt.imshow(hist_img,                         #numpy array generating the image
           cmap = 'gray')   #algorithm used to blend square colors; with 'nearest' colors will not be blended
          
plt.show()

print(hist_img)
#len(hist_img)
type(hist_img)
```

```python
image = img_as_float(data.chelsea()) #astronaut, chelsea, rocket, retina, logo(4 canales), coffee
print(image.shape)

fig=plt.figure()
plt.imshow(image)  
          
plt.show()
```

```python
ch0= image[2][0:len(image), 0]
hist_img_0,bins =exposure.histogram(ch0, 10)

fig=plt.figure(figsize=(12,4.5))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)

ax1.imshow(exposure.histogram(ch0, 10))
ax2.plot(bins, hist_img_0 / hist_img_0.max())
          
plt.show()
```
