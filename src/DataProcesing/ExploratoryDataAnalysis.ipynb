```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_map = pd.read_csv('/data/images_gz2/gz2_filename_mapping.csv')
df_classes = pd.read_csv('/data/images_gz2/gz2_hart16.csv')
df_classes = df_classes[TYPES]
```

```python

for type in TYPES:
    fig = plt.figure()
    plt.title(label=type, fontsize=15, color="green")
    plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)
    df_classes[type][df_classes[type] >.2].hist(bins=16, figsize=(12, 10))
    fig.savefig("/data/plots/histogram_galaxy_type_02/"+type+".jpg", bbox_inches='tight')
```
